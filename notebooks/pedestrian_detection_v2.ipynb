```python


import glob
import imutils
import cv2
import sys
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image, display
######################################

def normalize_grayimage(image):
	image = cv2.equalizeHist(image)
	#cv2.imshow("Equalized img", image)
	
	return image

#Values for statistical evaluation
num_ppl = 0
num_pos_detections = 0
num_false_positivos = 0 
num_false_negatives = 0

#IMG PATHS
images = glob.glob("img/*.jpg")
# cap = cv2.VideoCapture('/outputs/jitter-10/clips/105-0-0-0.mp4')
cap = cv2.VideoCapture('/outputs/jitter-10/clips/3-0-0-0.mp4')
cascPath = "haarcascade_pedestrian.xml"
pplCascade = cv2.CascadeClassifier(cascPath)

counter = 0
while cap.isOpened():
    # Reading the video stream
    ret, image = cap.read()

    if ret:
        image = imutils.resize(image,
                            width=min(image.shape[1]//6, image.shape[1]//6))
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        #gray = normalize_grayimage(gray)
            
        pedestrians = pplCascade.detectMultiScale(
            gray,
            scaleFactor=1.2,
            minNeighbors=2,
            minSize=(30,30),
            flags = cv2.CASCADE_SCALE_IMAGE
        )

        print("Found {0} ppl!".format(len(pedestrians)))
        
        
        
        #Draw a rectangle around the detected objects
        for (x, y, w, h) in pedestrians:
            cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
        display(Image(data=cv2.imencode('.jpg', image)[1].tostring()))
		# Showing the output Image
		# cv2.imshow("Image", image)
        
        if counter == 120:
            break
        
        counter += 1

        # outputname = "testoutput/output_"+filename.split("/")[1]
        # cv2.imwrite(outputname, image)
        #cv2.imshow("Ppl found", image)
        #cv2.waitKey(0)

cap.release()
```
